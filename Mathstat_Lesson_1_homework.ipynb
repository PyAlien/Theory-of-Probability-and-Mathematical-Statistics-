### ***1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.*** 

#### a) Найти вероятность того, что все карты – крести. 

Событие А = {извлечены четыре карты - трефы}

```python
from math import factorial
```

```python
# Функция для вычисления числа сочетаний
def combinations(n: int, k: int) -> int: 
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

***Первый способ решения с помощью сочетаний***

```python
# Найдем число благоприятных исходов равное числу способов, которыми можно извлечь 4 трефы из 13 карт этой масти
# n = 13 - число всех возможных исходов, где все карты - трефы 
# m = 4 - число исходов, благоприятствующих появлению четырех карт - треф

print(f'{combinations(13, 4)} сочетаний, где все четыре карты будут - трефы')
```

```python
# Найдем общее число всех возможных элементарных исходов испытания 
# равное числу способов, которыми можно извлечь 4 карты из колоды 52 карты
# n = 52 - число всех возможных исходов из всех карт
# m = 4 - число исходов, благоприятствующих появлению четырех карт

print(f'{combinations(52, 4)} сочетаний из четырех карт')
```

```python
# Найдем вероятность того, что четыре карты - трефы
# n = 270725 число всех возможных сочетаний из четырех карт
# m = 715 число возможных сочетаний, где все четыре карты будут - трефы
# (m / n) * 100 - Вероятность того, что все карты - трефы

print(f'Вероятность того, что все четыре карты - трефы = {round((715 / 270725) * 100, 2)}%')
```

***Второй способ с помощью произведения вероятностей***

```python
# Карты достают из колоды последовательно и не возвращая их обратно в колоду
# Значит имеем дело с условной вероятностью

print(f'Вероятность того, что первая карта - трефы = {round((13 / 52) * 100, 2)}%')
print(f'Вероятность того, что вторая карта - трефы = {round((12 / 51) * 100, 2)}%')
print(f'Вероятность того, что третья карта - трефы = {round((11 / 50) * 100, 2)}%')
print(f'Вероятность того, что четвертая карта - трефы = {round((10 / 49) * 100, 2)}%')

# Вероятность того, что все четыре карты - трефы равна произведению вероятностей событий выше

print(f'Вероятность того, что все четыре карты - трефы =\
      {round((13 / 52) * (12 / 51) * (11 / 50) * (10 / 49) * 100, 2)}%')
```

#### б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Событие А = {извлечены четыре карты - как минимум одна из них туз}.<p> Событие противоположное А = {извлечены четыре карты - все не тузы}

***Первый способ решения с помощью сочетаний***

```python
# Найдем число благоприятных исходов равное числу способов, которыми можно извлечь 4 карты из 48 карт без тузов
# n = 48 - число всех возможных исходов, где нет тузов 
# m = 4 - число исходов, благоприятствующих появлению четырех карт

print(f'{combinations(48, 4)} сочетаний, где нет тузов')
```

```python
# Найдем общее число всех возможных элементарных исходов испытания 
# равное числу способов, которыми можно извлечь 4 карты из колоды 52 карты
# n = 52 - число всех возможных исходов из всех карт
# m = 4 - число исходов, благоприятствующих появлению четырех карт

print(f'{combinations(52, 4)} сочетаний из четырех карт')
```

```python
# Найдем вероятность того, что четыре карты - не тузы
# n = 270725 число всех возможных сочетаний из четырех карт
# m = 194580 число возможных сочетаний, где все четыре карты будут - не тузы
# 1 - (m / n) * 100 - Вероятность того, что как минимум одна карта из четырех - туз

print(f'Вероятность того, что все четыре карты - трефы = {round((1 - (194580 / 270725)) * 100, 2)}%')
```

***Второй способ с помощью произведения вероятностей***

```python
# Карты достают из колоды последовательно и не возвращая их обратно в колоду
# Значит имеем дело с условной вероятностью

print(f'Вероятность того, что первая карта - не туз = {round((48 / 52) * 100, 2)}%')
print(f'Вероятность того, что вторая карта - не туз = {round((47 / 51) * 100, 2)}%')
print(f'Вероятность того, что третья карта - не туз = {round((46 / 50) * 100, 2)}%')
print(f'Вероятность того, что четвертая карта - не туз = {round((45 / 49) * 100, 2)}%')

# Вероятность того, что как минимум одна карта из четырех - туз

print(f'Вероятность того, что как минимум одна карта из четырех - туз =\
      {round((1 - ((48 / 52) * (47 / 51) * (46 / 50) * (45 / 49))) * 100, 2)}%')
```

***Первый способ решения с помощью сочетаний***

```python
# Найдем общее число всех возможных элементарных исходов испытания 
# равное числу способов, которыми можно подобрать 3 кнопки из 10 кодового замка
# n = 10 - число всех возможных исходов равное 10 кнопкам
# m = 3 - число исходов, благоприятствующих появлению нужной комбинации

print(f'{combinations(10, 3)} сочетаний комбинаций трех кнопок')
```

```python
# Число сочетаний благоприятных исходов из подмножества кнопок задействованных в комбинации равно 1
# Найдем вероятность того, что человек, не знающий код, откроет дверь с первой попытки 

print(f'Вероятность того, что человек, не знающий код, откроет дверь с первой попытки =\
      {round((1 / 120) * 100, 2)}%')
```

### ***3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?***

***Первый способ решения с помощью сочетаний***

```python
# Найдем число благоприятных исходов равно числу способов, которыми можно извлечь 3 детали из 9 окрашенных:
# n = 9 - число всех возможных исходов равное девяти окрашенным деталям
# m = 3 - число исходов, благоприятствующих появлению трех окрашенных деталей

print(f'{combinations(9, 3)} сочетания трех окрашенных деталей')
```

```python
# Найдем общее число всех возможных элементарных исходов испытания 
# равное числу способов, которыми можно извлечь 3 детали из 15
# n = 15 - число всех возможных исходов из всех деталей
# m = 3 - число исходов, благоприятствующих появлению 3 окрашенных деталей

print(f'{combinations(15, 3)} сочетаний из всех деталей')
```

```python
# Найдем вероятность того, что все извлеченные детали окрашены

print(f'Вероятность того, что все извлеченные детали окрашены =\
      {round((84 / 455) * 100, 2)}%')
```

***Второй способ с помощью произведения вероятностей***

```python
# Детали извлекают последовательно и обратно не возвращают
# Значит имеем дело с условной вероятностью

print(f'Вероятность того, что первая деталь окрашена = {round((9 / 15) * 100, 2)}%')
print(f'Вероятность того, что вторая кнопка совпала = {round((8 / 14) * 100, 2)}%')
print(f'Вероятность того, что третья кнопка совпала = {round((7 / 13) * 100, 2)}%')

# Вероятность того, что все извлеченные детали окрашены

print(f'Вероятность того, что все извлеченные детали окрашены =\
      {round(((9 / 15) * (8 / 14) * (7 / 13)) * 100, 2)}%')
```

### ***4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?***

***Первый способ решения с помощью сочетаний***

```python
# Найдем число возможных сочетаний двух билетов из ста билетов
# n = 100 - число всех возможных исходов равное количеству билетов
# m = 2 - число исходов, благоприятствующих появлению двух любых билетов

print(f'{combinations(100, 2)} сочетаний двух билетов из ста')
```

```python
# Число сочетаний благоприятных исходов из подмножества выигрышных билетов равно одному
# Найдем вероятность того, что два приобретенных билета окажутся выигрышными

print(f'Вероятность того, что 2 приобретенных билета окажутся выигрышными =\
      {round((1 / 4950) * 100, 2)}%')
```

***Второй способ с помощью произведения вероятностей***

```python
# Билеты приобретают последовательно и обратно не возвращают
# Значит имеем дело с условной вероятностью

print(f'Вероятность того, что первый билет выигрышный = {round((2 / 100) * 100, 2)}%')
print(f'Вероятность того, что второй билет выигрышный = {round((1 / 99) * 100, 2)}%')

# Вероятность того, что 2 приобретенных билета окажутся выигрышными

print(f'Вероятность того, что 2 приобретенных билета окажутся выигрышными =\
      {round(((2 / 100) * (1 / 99)) * 100, 2)}%')
```
