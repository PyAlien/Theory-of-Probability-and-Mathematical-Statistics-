### ***1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.***

```python
import numpy as np
```

```python
mean = (800 + 200) / 2
print(f'Среднее значение случайной непрерывной величины А = {mean}')
```

```python
variance = np.power((800 - 200), 2) / 12
print(f'Дисперсия случайной непрерывной величины А = {variance}')
```

### ***2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.***

Зная дисперсию и левую границу случайной непрерывной равномерно распределенной величине B
можно найти правую границу и ее среднее значение.

```python
variance = 0.2
a = 0.5
b = np.sqrt(variance * 12) + a
print(f'Правая граница = {b}')
print(f'Среднее значение = {(a + b) / 2}')
```

### ***4. Рост взрослого населения города X имеет нормальное распределение.***
    Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
    Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
    а). больше 182 см
    б). больше 190 см
    в). от 166 см до 190 см
    г). от 166 см до 182 см
    д). от 158 см до 190 см
    е). не выше 150 см или не ниже 190 см
    ё). не выше 150 см или не ниже 198 см
    ж). ниже 166 см.

```python
mean_height = 174
dev_height = 8
```

```python
# a) рост больше 182 см, т.е. больше одного стандартного отклонения
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z = (182 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z}')
# Т.к. z=1=стандартному отклонению(s)
# В пределах 1s находится 68% данных
print(f'Вероятность того, что человек имеет рост больше 182 см равен {(100 - 68) / 2}%')
```

```python
# б) рост больше 190 см, т.е. больше 2s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z = (190 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z}')
# Т.к. z=2=2s
# В пределах 2s находится 95% данных
print(f'Вероятность того, что человек имеет рост больше 190 см равен {(100 - 95) / 2}%')
```

```python
# в) от 166 см до 190 см, т.е. от -s до 2s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z1 = (166 - mean_height) / dev_height
z2 = (190 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z1, z2}')
# Т.к. z=-1=-s и 2s
# В пределах 1s находится 68% данных
# В пределах 2s находится 95% данных
print(f'Вероятность того, что человек имеет рост от 166 см до 190 см равен\
      {((95 - 68) / 2) + 68}%')
```

```python
# г) от 166 см до 182 см, т.е. от -s до s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z1 = (166 - mean_height) / dev_height
z2 = (182 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z1, z2}')
# Т.к. z1=-1=-s и z2=1=s
# В пределах 1s находится 68% данных
print(f'Вероятность того, что человек имеет рост от 166 см до 182 см равен 68%')
```

```python
# д) от 158 см до 190 см, т.е. от -2s до 2s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z1 = (158 - mean_height) / dev_height
z2 = (190 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z1, z2}')
# Т.к. z1=-2=-2s и z2=2=2s
# В пределах 2s находится 95% данных
print(f'Вероятность того, что человек имеет рост от 158 см до 190 см равен 95%')
```

```python
# е) не выше 150 см или не ниже 190 см, т.е.ниже -3s и выше 2s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z1 = (150 - mean_height) / dev_height
z2 = (190 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z1, z2}')
# Т.к. z1=-2=-3s и z2=2=2s
# В пределах 2s находится 95% данных
# В пределах 3s находится 99.7% данных
print(f'Вероятность того, что человек имеет рост не выше 150 см или не ниже 190 см равен\
        {round((100 - 95) - ((99.7 - 95) / 2), 3)}%') 
```

```python
# ё) не выше 150 см или не ниже 198 см, т.е.ниже -3s и выше 3s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z1 = (150 - mean_height) / dev_height
z2 = (198 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z1, z2}')
# Т.к. z1=-3=-3s и z2=3=3s
# В пределах 3s находится 99.7% данных
print(f'Вероятность того, что человек имеет рост не выше 150 см или не ниже 198 см равен\
        {round(100 - 99.7, 3)}%')
```

```python
# ж) ниже 166 см, т.е.ниже -s
# Переход от нормального распределения величины роста к величине
# со стандартным нормальным рапределением z
z = (166 - mean_height) / dev_height
print(f'Величина со стандартным нормальным распределением = {z}')
# Т.к. z=-1=-1s
# В пределах 1s находится 68% данных
# В пределах 2s находится 95% данных
# В пределах 3s находится 99.7% данных
print(f'Вероятность того, что человек имеет рост ниже 166 см равен {(100 - 68) / 2}%')
```
