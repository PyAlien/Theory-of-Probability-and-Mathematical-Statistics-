### ***1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.*** 

```python
import numpy as np
from math import factorial
```

```python
p = 0.8 # Вероятность попадания в мишень с первого раза
n = 100 # Количество выстрелов
m = 85 # Число попаданий в мишень

lambda_ = n * p # среднее количество событий за фиксированный промежуток времени 

print(f'Вероятность того, что стрелок попадет в цель ровно 85 раз равна\
        {round(((np.power(lambda_, m) / factorial(m)) * np.exp(-lambda_)) * 100, 3)}%')
```

### ***2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?*** 

```python
p = 0.0004 # Вероятность того, что лампочка перегорит в течение первого дня эксплуатации
n = 5000 # Количество включенных лампочек
m1 = 0 # Количество перегоревших лампочек в первом случае
m2 = 2 # Количество перегоревших лампочек во втором случае

lambda_ = n * p # интенсивность, с которой события происходят

print(f'Вероятность, что ни одна из них не перегорит в первый день равна\
        {round(((np.power(lambda_, m1) / factorial(m1)) * np.exp(-lambda_)) * 100, 3)}%')
print(f'Вероятность, что перегорят ровно две равна\
        {round(((np.power(lambda_, m2) / factorial(m2)) * np.exp(-lambda_)) * 100, 3)}%')
```

### ***3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?*** 

```python
p = 0.5 # Вероятность того, что выпадет орёл
n = 144 # Количество испытаний
m = 70 # Количество испытаний, когда выпал орёл

lambda_ = n * p # интенсивность, с которой событие происходит

print(f'Вероятность, что орёл выпадет ровно 70 раз равна\
        {round(((np.power(lambda_, m) / factorial(m)) * np.exp(-lambda_)) * 100, 3)}%')
```

### ***4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?***

```python
# Функция для вычисления числа сочетаний

def combinations(k: int, n: int) -> int: 
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Функция для вычисления гипергеометрической вероятности

def hypergeometric(K: int, N: int, n: int, k: int) -> int:
    return round((combinations(k, K) * combinations((n - k), (N - K))) / combinations(n, N), 2)
```

```python
N1 = 10 # Количество мячей в первом ящике
K1 = 7 # Количество белых мячей в первом ящике
N2 = 11 # Количество мячей во втором ящике
K2 = 9 # Количество белых мячей во втором ящике
n1 = 2 # Количество мячей, извлеченных из первого ящика
n2 = 2 # Количество мячей, извлеченных из второго ящика
```

```python
# a) Какова вероятность того, что все мячи белые?
# События:

k1 = 2 # Количество белых мячей, извлеченных из первого ящика
k2 = 2 # Количество белых мячей, извлеченных из второго ящика

# Первый способ - задача на гипергеометрическое распределение

p_1 = hypergeometric(7, 10, 2, 2)
p_2 = hypergeometric(9, 11, 2, 2)

print(f'Вероятность того, что все мячи белые равна\
        {round(p_1 * p_2 * 100, 2)}%')

# Второй способ - задача на условную вероятность

print(f'Вероятность того, что все мячи белые равна\
        {round((((K1 / N1) * ((K1 - 1) / (N1 - 1))) * ((K2 / N2) * ((K2 - 1) / (N2 - 1)))) * 100, 2)}%')
```

```python
# б) Какова вероятность того, что ровно два мяча белые?

# Первый случай (A)

k1 = 2 # Количество белых мячей, извлеченных из первого ящика
k2 = 0 # Количество белых мячей, извлеченных из второго ящика
p_1_case_A = hypergeometric(7, 10, 2, 2)
p_2_case_A = hypergeometric(9, 11, 2, 0)
multi_case_A = p_1_case_A * p_2_case_A

print(p_1_case_A, p_2_case_A)

# Второй случай (B)

k1 = 0 # Количество белых мячей, извлеченных из первого ящика
k2 = 2 # Количество белых мячей, извлеченных из второго ящика
p_1_case_B = hypergeometric(7, 10, 2, 0)
p_2_case_B = hypergeometric(9, 11, 2, 2)
multi_case_B = p_1_case_B * p_2_case_B

print(p_1_case_B, p_2_case_B)

# Третий случай (С)

k1 = 1 # Количество белых мячей, извлеченных из первого ящика
k2 = 1 # Количество белых мячей, извлеченных из второго ящика
p_1_case_C = hypergeometric(7, 10, 2, 1)
p_2_case_C = hypergeometric(9, 11, 2, 1)
multi_case_C = p_1_case_C * p_2_case_C

print(p_1_case_C, p_2_case_C)

print(f'Вероятность того, что ровно два мяча белые равна {(multi_case_A + multi_case_B + multi_case_C) * 100}%')
```
